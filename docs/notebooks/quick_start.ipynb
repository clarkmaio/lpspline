# LPSpline Demo

This notebook demonstrates how to use the lpspline package to build and optimize linear spline models using the additive API.

```
import numpy as np
import polars as pl
from lpspline import l, pwl, bs, cs, f
from lpspline.viz import plot_diagnostic
from lpspline.datasets import load_demo_dataset

# Set seed for reproducibility
np.random.seed(42)
```

## 1. Generate Synthetic Data

We create a dataset with various non-linear relationships.

```
X, y = load_demo_dataset(samples=1000)
```

## 2. Define Model

We define the components and add them together to create the model.

```
model = (
    +l(term='xl', by='xfactor')
    + pwl(term='xpwl', knots=3)
    + bs(term="xbs", knots=10, degree=2)
    + cs(term="xcyc", order=3)
    + f(term="xfactor")
)
```

## 3. Fit the Model

We use the fit method on the combined model.

```
model.fit(X, y)
```

## 4. Visualization

Visualize the fitted components.

```
plot_diagnostic(model=model, X=X, y=y, ncols=3)
```
